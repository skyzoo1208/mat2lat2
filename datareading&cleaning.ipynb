```python
#-- Merubah male menjadi 1 atau pria --
latih.loc[latih['Sex']=='male','Sex'] = 1
#-- Merubah Female menjadi 0 atau wanita --
latih.loc[latih['Sex']=='female','Sex'] = 0
```

```python
#-- melihat data latih yang Survived (selamat)
latih['Survived'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(2, 2))
#-- data adalah nama_file, x adalah field --
sns.countplot(data=latih, x='Survived', ax=ax)
```
